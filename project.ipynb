# Importing The Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Creating dataset

```python
data = np.random.randint(0, 100, size=208)
data2 = np.random.randint(0, 10, size=208)
print(data)
print(data2)
```

# Giving irregular values to the dataset

```python
data[0] = 2000
data[10] = -139
data[67] = 0
data[59] = 786
data[92] = -456
data[0] = 2679
data[10] = 139
data[67] = 0
data[59] = -886
data[92] = -456
data2[0] = -7899
data2[12] = 56
data2[69] = -99
data2[62] = 678
data2[94] = 0
data2[0] = -999
data2[12] = 560
data2[69] = -99
data2[62] = -678
data2[94] = 89
```

# Creating The dataframe

```python
df = pd.DataFrame()
df['Average hours studied'] = (data2)
df['average percentage scored'] = (data)

df
```

# Removing erraneous data using predefined functions

```python
df.isnull().sum()
for x in df.index:
    if df.loc[x,'Average hours studied'] > 10:
        df.drop(x, inplace = True)
for x in df.index:
    if df.loc[x,'Average hours studied'] < 0:
        df.drop(x, inplace = True)
for x in df.index:
    if df.loc[x,'average percentage scored'] > 100:
        df.drop(x, inplace = True)
for x in df.index:
    if df.loc[x,'average percentage scored'] < 0:
        df.drop(x, inplace = True)
```
